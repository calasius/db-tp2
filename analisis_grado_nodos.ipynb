```python
import pandas as pd
```

```python
degs = pd.read_csv("in_out_degree.csv")
degs.columns = ['in_degree','out_degree']
```

```python
%matplotlib inline
degs.hist(bins=10,figsize=(20,20))
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(degs)
degs_norm = pd.DataFrame(np_scaled)
degs_norm.columns= ['in_degree','out_degree']
degs_norm.hist(bins=100)
```
